Method I - Calculate the bootstrap ourselves using a "for-loop" and resampling the indices.  Additionally, we will vary (or permute) the order of the predictors on each model estimation

We will estimate bootstrap confidence intervals using two different approaches.  Both methods will 
need a function to be defined that calculates the regression coefficients and returns them for accumulation

```r
# The code segment that we'll need to construct the formula from one model estimation to the 
# next model estimation;  Notice how we are constructing a permuted variable ordering for the 
# model to estimate
paste(sample(names(x_df)[-1], ncol(x_df)-1, replace=FALSE), collapse="+")
```

```r
# Define the function that will be called many times; "boot_samps" will determine how 
# how many bootstrap samples to take;  This will be 1000 for our purposes.
# We create a calling argument "permute_vars", that will allow us to use this function
# for Method I and Method II when constructing the bootstrap CI

lm_function<-function(data, i, col_index, permute_vars=FALSE){
   col_index=1
   if(permute_vars==TRUE){
       lm_fit_coefs<-lm(as.formula(paste(names(data)[col_index],"~",
                                   paste(sample(names(data)[-1], 
                                                ncol(x_df)-1, replace=FALSE), collapse="+"))), 
                        data)$coefficients
    }
    if(permute_vars==FALSE){
       data<-data[i, ] 
       lm_fit_coefs<-lm(var1 ~ ., data)$coefficients     
    }
   return(lm_fit_coefs)
}
```

```r
# Begin the main looping for Method I construction of bootstrap CI
# Initialize variables
col_index<-1
boot_samps<-1000
coef_mat_accum<-matrix(NA, ncol=ncol(x_df), nrow=boot_samps)

# Get the coefficients and the coefficient names
# Note that the returning coefficients are of type "vector"
coef_out<-lm(var1 ~. , data=x_df)$coefficients
is.vector(coef_out)
names(coef_out)

# Order the coefficient names and reassign the names attribute 
# back to the matrix object;  Note that the vector has a "names"
# attribute for the row names;  but a "matrix" has a "colnames" attribute
# for it's columns; this is why the right side uses "names", but the 
# column names assignment ot our matrix uses "colnames"
colnames(coef_mat_accum)<-names(coef_out)[order(names(coef_out))]

# Display names
colnames(coef_mat_accum)

# Bootstrap sampling - using for-loop and sampling with replacement (SWR) 
for(i in 1:boot_samps){
    print(i)
    boot_index<-sample(1:nrow(x_df), 10, replace=TRUE)
    x_df_rep<-x_df[boot_index, ]
    coef_out<-lm_function(data=x_df_rep, col_index=col_index, permute_vars=TRUE)
    coef_out<-coef_out[order(names(coef_out))]
    coef_mat_accum[i, ] <- coef_out 
    print(coef_mat_accum[i,1])
    }
```

```r
# Calculate the quantiles of each column;  These are the CI:  [LL_beta_.025,  UL_beta_.975]
# These betas define an uncertainty interval that capture 95% of the variation in the sample's betas
apply(coef_mat_accum, 2, quantile, probs=c(.025, .975), na.rm=TRUE)
```

Method II - Calculate bootstrap confidence intervals using the R package "boot"

```r
# Bootstrap analysis using package "boot" -  We DON'T want to permute the predictors
# when using package "boot";  The bootstrap averaging of coefficients occurs across 
# models where the ordering of the predictors is the same for each bootstrap sample.
# Notice var10 will not be able to be estimated;  n must be LARGER than p to get estimates
# for all 10 variables 

library(boot)
boot.fit<-boot(data=x_df, lm_function, R=1000, col_index=1, permute_vars=FALSE)
boot.fit
```

```r
# Using the function "boot.ci" from package "boot", calculate 
# bootstrap CI for each predictor (including the intercept).
# Note that var10 will be missing for every model;  this variable is the 
# Last predictor in the model

# List the coefficient names 
names(boot.fit$t0)

# Print the bootstrap CI
for(j in 1:10){
    cat("\n")
    cat("\n")
    print(names(boot.fit$t0[j]))
    cat("\n")
    cat("\n")
    print(boot.ci(boot.fit,  t0 = boot.fit$t0[j], t = boot.fit$t[,j], type="perc"))
}
```

```r
# Notice that the last predictor in the sorted vector of predictors is "var9"
# and it consists of all NA's;  The point here is that Method II which uses the SAME
# model for each bootstrap sample, CANNOT estimate ALL of the cofficients and their standard
# errors;  But Method I, which uses permutations of the predictors, or each model estimation, 
# DOES ALLOW every predictor to be estimated a number of times - enough times to get a rough 
# estimate of the average coefficient value and its standard devaition (the standard error)
print(summary(boot.fit$t[,10]))
```
